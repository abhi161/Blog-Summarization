## Importing libraries

```python
from transformers import pipeline
from bs4 import BeautifulSoup
import requests
```

```python
summarize = pipeline("summarization")
```

# Scrapping the Blog

```python
URL = "https://hackernoon.com/developing-ai-security-systems-with-edge-biometrics"

blog = requests.get(URL)

blog.text
```

```python
soup = BeautifulSoup(blog.text,"html.parser")
article = soup.findAll(['h1','p'])
```

```python
article = [text.text for text in article]
```

```python
article = ' '.join(article)
article
```

**Dividing into chunks:-** 
Since Summarization pipeline hence word limit or sentence limit to be processed at once (1024) we have to divide the complete paragraph into tokens / chunks


```python
article = article.replace('.','.<eos>')
article = article.replace('?','?<eos>')
article = article.replace('!','!<eos>')
sentences = article.split('<eos>')

```

```python
sentences
```

```python
chunks=  []
max_size =500
chunk_counter = 0

for sentence in sentences:
  if len(chunks) == chunk_counter +1:
    if len(chunks[chunk_counter]) + len(sentence.split(' ')) <= max_size:
      chunks[chunk_counter].extend(sentence.split(' '))
    else:
      chunk_counter+=1
      chunks.append(sentence.split(' '))
  else:
    chunks.append(sentence.split(' '))

chunks
```

```python
for chunk_id in range(len(chunks)):
  chunks[chunk_id] = ' '.join(chunks[chunk_id])
```

```python
chunks
```

```python
len(chunks[0].split(' '))
```

## Model Creation

```python
res= summarize(chunks, max_length=100, min_length =30,do_sample=False)
res
```

```python
Summary = ' '.join(text['summary_text'] for text in res)
Summary
```
